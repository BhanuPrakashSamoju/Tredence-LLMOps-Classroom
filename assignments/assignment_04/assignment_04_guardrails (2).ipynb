```python
new_input_scanners = [
    Anonymize(vault,recognizer_conf=BERT_LARGE_NER_CONF, language="en",use_faker='[Not Allowed]'), 
    BanTopics(topics=["salaries"], threshold=0.5), 
    PromptInjection(threshold=0.75, match_type=PromptInjectionMatchType.FULL)
]

def get_guardrailed_response(prompt):
    sanitized_prompt1, results_valid1, results_score1 = scan_prompt(new_input_scanners, prompt)
    if all(results_valid1.values()) is False:
        print(f"Prompt {prompt} is not valid, scores: {results_score}")
        return f"I am sorry, I can't help with your request because of failed prompt validation. Kindly ask again. \n{results_valid1}"
    else:
        response = call_LLM(prompt)
        return response
```

```python
import asyncio

async def validate_guardrails(prompt):
    sanitized_prompt1, results_valid1, results_score1 = scan_prompt(new_input_scanners, prompt)
    if all(results_valid1.values()) is False:
        print(f"I am sorry, I can't help with your request because of failed prompt validation. Kindly ask again. \n{results_valid1}")
        return ("Fail", results_valid1, results_score1, sanitized_prompt1)
    else:
        return ("Pass", results_valid1, results_score1, sanitized_prompt1)

async def get_chat_response(p):
    try:
        response = call_LLM(p)
        return response
    except Exception as e:
        print(f"Ran into exception: {e}")
```

```python
async def execute_chat_with_guardrail(user_request):
    guardrail_task = asyncio.create_task(validate_guardrails(user_request))
    chat_task = asyncio.create_task(get_chat_response(user_request))

    while True:
        done, _ = await asyncio.wait(
            [guardrail_task, chat_task], return_when=asyncio.FIRST_COMPLETED
        )
        if guardrail_task in done:
            guardrail_response = guardrail_task.result()
            if guardrail_response[0] == "Fail":
                chat_task.cancel()
                print("guardrail triggered")
                return f"I am sorry, I can't help with your request because of failed prompt validation. Kindly ask again. \n{guardrail_response}"
            elif chat_task in done:
                chat_response = chat_task.result()
                return chat_response
        else:
            await asyncio.sleep(0.1)  # sleep for a bit before checking the tasks again
```

```python
await execute_chat_with_guardrail(p1)
```
