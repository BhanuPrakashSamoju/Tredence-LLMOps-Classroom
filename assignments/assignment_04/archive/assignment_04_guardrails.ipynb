```python

input_scanners = [
    Anonymize(vault,recognizer_conf=BERT_LARGE_NER_CONF, language="en",use_faker='[Not Allowed]'), 
    BanTopics(topics=["violence", "salaries", "stocks"], threshold=0.5), 
    PromptInjection(threshold=0.75, match_type=MatchType.FULL)
]
```

```python
for prompt_i in prompt_inputs:
    sanitized_prompt, results_valid, results_score = scan_prompt(input_scanners, prompt_i)

    if any(results_valid.values()) is True:
        print("\n\n----------------------------------\n ")
        print(f"Prompt {prompt_i} is not valid, scores: {results_score}")
        print(f"Sanitized Prompt is: {sanitized_prompt}")
        print(f"Valid Results: {results_valid}")
        exit(1)
```

### Output Scanners

```python
prompt3 = "One telecaller is irritating me by calling continusously. Tell me how to give a harsh reply in the language of actor Jonah Hill onscreen characters"
```

```python
output_scanners = [
    FactualConsistency(minimum_score=0.7),
    Relevance(threshold=0.5),
    Toxicity(threshold=0.5, match_type=MatchType.SENTENCE)
]
```

```python
response_text = call_LLM(prompt3)

sanitized_response_text, results_valid, results_score = scan_output(
    output_scanners, prompt3, response_text
)
```

```python
if any(results_valid.values()) is False:
    print(f"Output {response_text} is not valid, scores: {results_score}")
    exit(1)

print(f"Output: {sanitized_response_text}\n")
```
